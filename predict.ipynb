```python
import tensorflow._api.v2.compat.v1 as tf
import numpy as np
from pandas.io.parsers import read_csv
#tf.disable_v2_behavior()
tf.compat.v1.disable_eager_execution()
tf.compat.v1.disable_v2_behavior()
```

```python
tf.global_variables_initializer()
data = read_csv('price data.csv', sep=',')
xy=np.array(data, dtype=np.float32)
x_data = xy[:, 1:-1]
y_data = xy[:, [-1]]
```

```python
x = tf.placeholder(tf.float32, shape=[None, 4])
y = tf.placeholder(tf.float32, shape=[None, 1])
w=tf.Variable(tf.random_normal([4, 1]), name="weight")
b=tf.Variable(tf.random_normal([1]), name="bias")
```

```python
hypo = tf.matmul(x, w) + b
cost = tf.reduce_mean(tf.square(hypo - y))
```

```python
optimizer = tf.train.GradientDescentOptimizer(learning_rate = 0.000005)
train=optimizer.minimize(cost)
```

```python
sees = tf.Session()
sees.run(tf.global_variables_initializer())
```

```python
for i in range(100001):
    cost_, hypo_, _ = sees.run([cost, hypo, train], feed_dict={x:x_data, y:y_data})
    if i%500 ==0:
        print("#", i, "손실비용: ", cost_)
        print(i, "배추가격: ", hypo_[0])
    
```

```python
saver = tf.train.Saver()
save_path =saver.save(sees, "./saved.ckpt")
```

```python
test_arr=[[8, 4.9, 11.3, 10.8]]
predict=sees.run(hypo, feed_dict={x:test_arr})
print(predict)
```

```python
sess2 = tf.Session()
saver.restore(sess2, "./saved.ckpt")
predict = sess2.run(hypo, feed_dict={x:test_arr})
print(predict[0])
```
