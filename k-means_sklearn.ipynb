```python
from sklearn.cluster import KMeans
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
import numpy as np
import seaborn as sb
from random import *
import copy
```

```python
df = pd.DataFrame(columns=['x', 'y'])
df.loc[0] = [2, 3]
temp_data = [0, 0]
for i in range(30):
    temp_data[0] = randint(1, 20)
    temp_data[1] = randint(1, 20)
    df.loc[i]=copy.deepcopy(temp_data)
```

```python
df.head(30)
```

```python
sb.lmplot('x', 'y', data=df, fit_reg=False, scatter_kws={'s':100})
plt.title("k-means algorithm")
plt.xlabel('x')
plt.ylabel('y')
```

```python
points = df.values
kmeans = KMeans(n_clusters = 4).fit(points)
kmeans.cluster_centers_
```

```python
df['cluster'] = kmeans.labels_
df.head(30)
```

```python
sb.lmplot('x', 'y', data=df, fit_reg=False, scatter_kws={'s':150}, hue="cluster")
plt.title("k-means algorithm")
plt.xlabel('x')
plt.ylabel('y')
```
